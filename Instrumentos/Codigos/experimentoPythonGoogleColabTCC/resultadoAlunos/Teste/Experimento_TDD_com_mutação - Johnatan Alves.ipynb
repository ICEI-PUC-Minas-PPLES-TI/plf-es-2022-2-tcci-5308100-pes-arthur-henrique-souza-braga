# Segundo passo

> À esquerda da tela, localize o ícone da pasta e clique sobre ele. Em seguida, abra a classe "Livro" que acabamos de criar e cole o código a seguir.

```python
class Livro:
    def __init__(self, titulo, autor, ano_publicacao, disponivel):
        self.titulo = titulo
        self.autor = autor
        self.ano_publicacao = ano_publicacao
        self.disponivel = disponivel

    def get_titulo(self):
        return self.titulo

    def set_titulo(self, titulo):
        self.titulo = titulo

    def get_autor(self):
        return self.autor

    def set_autor(self, autor):
        self.autor = autor

    def get_ano_publicacao(self):
        return self.ano_publicacao

    def set_ano_publicacao(self, ano_publicacao):
        self.ano_publicacao = ano_publicacao + 1

    def is_disponivel(self):
        return self.disponivel

    def set_disponivel(self, disponivel):
        self.disponivel = disponivel - 1
```

# Quarto passo

> No lado esquerdo da tela, localize e clique no ícone de pasta. Em seguida, abra a classe 'test_livro' que acabamos de criar e cole o código a seguir. Se a classe 'test_livro' não aparecer imediatamente, clique no ícone da pasta com um círculo para atualizar a lista de arquivos.

```python
import unittest
from livro import Livro

class TestLivro(unittest.TestCase):

    def test_get_titulo(self):
        livro = Livro("Título do Livro", "Autor do Livro", 2023, True)
        self.assertEqual(livro.get_titulo(), "Título do Livro")

    def test_set_titulo(self):
        livro = Livro("Título Antigo", "Autor do Livro", 2023, True)
        livro.set_titulo("Novo Título")
        self.assertEqual(livro.get_titulo(), "Novo Título")

    def test_get_autor(self):
        livro = Livro("Título do Livro", "Autor do Livro", 2023, True)
        self.assertEqual(livro.get_autor(), "Autor do Livro")

    def test_set_autor(self):
        livro = Livro("Título do Livro", "Autor Antigo", 2023, True)
        livro.set_autor("Novo Autor")
        self.assertEqual(livro.get_autor(), "Novo Autor")

    def test_get_ano_publicacao(self):
        livro = Livro("Título do Livro", "Autor do Livro", 2023, True)
        self.assertEqual(livro.get_ano_publicacao(), 2023)

    def test_set_ano_publicacao(self):
        livro = Livro("Título do Livro", "Autor do Livro", 2023, True)
        livro.set_ano_publicacao(2022)
        self.assertEqual(livro.get_ano_publicacao(), 2022)

    def test_is_disponivel(self):
        livro = Livro("Título do Livro", "Autor do Livro", 2023, True)
        self.assertTrue(livro.is_disponivel())

    def test_set_disponivel(self):
        livro = Livro("Título do Livro", "Autor do Livro", 2023, True)
        livro.set_disponivel(False)
        self.assertFalse(livro.is_disponivel())
```
